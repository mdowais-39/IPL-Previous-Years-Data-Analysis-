## batter

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df_batter = pd.read_csv("/content/Final_batter_ipl.csv")

# Preview
print("üßæ Data Shape:", df_batter.shape)
print("\nüîç First 5 Rows:\n", df_batter.head())
print("\nüßº Missing Values:\n", df_batter.isnull().sum())
print("\nüìä Data Types:\n", df_batter.dtypes)

# Summary statistics
print("\nüìà Descriptive Stats:\n", df_batter.describe())

# Distribution plots
plt.figure(figsize=(14, 4))

plt.subplot(1, 3, 1)
sns.histplot(df_batter['Runs'], kde=True, bins=15, color='skyblue')
plt.title("Runs Distribution")

plt.subplot(1, 3, 2)
sns.histplot(df_batter['Avg'], kde=True, bins=15, color='salmon')
plt.title("Batting Average Distribution")

plt.subplot(1, 3, 3)
sns.histplot(df_batter['S/R'], kde=True, bins=15, color='limegreen')
plt.title("Strike Rate Distribution")

plt.tight_layout()
plt.show()

# Correlation matrix
plt.figure(figsize=(6, 4))
sns.heatmap(df_batter[['Runs', 'Avg', 'S/R']].corr(), annot=True, cmap='coolwarm')
plt.title("üìå Correlation Matrix")
plt.show()

# Top batters by performance
top_batters = df_batter.sort_values(by='Runs', ascending=False).head(10)
print("\nüèè Top 10 Run Scorers:\n", top_batters[['Player', 'Runs', 'Avg', 'S/R']])
```

## head to head


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df_h2h_full = pd.read_csv("/content/batter_vs_bowler_head_to_head_2024.csv")

# Basic info
print("üßæ Data Shape:", df_h2h_full.shape)
print("\nüîç First 5 Rows:\n", df_h2h_full.head())
print("\nüßº Missing Values:\n", df_h2h_full.isnull().sum())
print("\nüìä Data Types:\n", df_h2h_full.dtypes)

# Summary
print("\nüìà Descriptive Stats:\n", df_h2h_full.describe())

# High sample H2H matchups
df_h2h_full['encounters'] = df_h2h_full['dismissals'] + (df_h2h_full['total_runs'] / (df_h2h_full['strike_rate'] + 1)) * 100 / 120  # Approx balls faced
high_sample = df_h2h_full[df_h2h_full['encounters'] > 5].sort_values(by='dismissals', ascending=False).head(10)
print("\nüí£ Most Deadly Bowler vs Batter Combos (5+ encounters):\n", high_sample)

# Efficiency score
df_h2h_full['bowler_efficiency'] = df_h2h_full['dismissals'] / (df_h2h_full['total_runs'] + 1)
df_h2h_full['batter_domination'] = df_h2h_full['strike_rate'] / (df_h2h_full['dismissals'] + 1)

# Top 10 most dominant batters in H2H (by domination score)
top_dom_batters = df_h2h_full[df_h2h_full['total_runs'] >= 40].sort_values(by='batter_domination', ascending=False).head(10)
print("\nüî• Top 10 Dominant Batters in H2H (min 40 runs):\n", top_dom_batters[['batter', 'bowler', 'total_runs', 'strike_rate', 'batter_domination']])

# Top 10 bowler threats
top_threat_bowlers = df_h2h_full[df_h2h_full['dismissals'] >= 3].sort_values(by='bowler_efficiency', ascending=False).head(10)
print("\nüíÄ Top 10 Most Effective Bowlers in H2H:\n", top_threat_bowlers[['batter', 'bowler', 'dismissals', 'total_runs', 'bowler_efficiency']])
```

## bowler venue

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df_bowler_venue = pd.read_csv("/content/bowler_performance_by_venue_2023_2024.csv")

# Preview
print("üßæ Data Shape:", df_bowler_venue.shape)
print("\nüîç First 5 Rows:\n", df_bowler_venue.head())
print("\nüßº Missing Values:\n", df_bowler_venue.isnull().sum())
print("\nüìä Data Types:\n", df_bowler_venue.dtypes)

# Summary
print("\nüìà Descriptive Stats:\n", df_bowler_venue.describe())

# Top venues with highest total wickets (to spot spin/pace bias)
venue_wickets = df_bowler_venue.groupby("venue")['wickets'].sum().sort_values(ascending=False).head(10)
print("\nüèüÔ∏è Venues with Most Wickets Taken:\n", venue_wickets)

# Best bowlers at most wicket-taking venues
top_venues = venue_wickets.index.tolist()[:3]
for venue in top_venues:
    print(f"\nüí• Top Bowlers at {venue}:\n")
    print(df_bowler_venue[df_bowler_venue['venue'] == venue].sort_values(by='wickets', ascending=False).head(5))

# Average economy rate per venue
plt.figure(figsize=(12, 5))
sns.barplot(x='venue', y='economy', data=df_bowler_venue, estimator='mean', ci=None, palette='coolwarm')
plt.xticks(rotation=45)
plt.title("üìâ Average Economy Rate per Venue")
plt.tight_layout()
plt.show()

# Create a venue performance score (optional feature for modeling)
df_bowler_venue['venue_score'] = df_bowler_venue['wickets'] / (df_bowler_venue['economy'] + 1)
```

## batter venue

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df_batter_venue = pd.read_csv("/content/batter_venue_stats_2023_2024.csv")

# Preview
print("üßæ Data Shape:", df_batter_venue.shape)
print("\nüîç First 5 Rows:\n", df_batter_venue.head())
print("\nüßº Missing Values:\n", df_batter_venue.isnull().sum())
print("\nüìä Data Types:\n", df_batter_venue.dtypes)

# Summary
print("\nüìà Descriptive Stats:\n", df_batter_venue.describe())

# Most dismissals at specific venues (tough for batters)
venue_dismissals = df_batter_venue.groupby("venue")['dismissals'].sum().sort_values(ascending=False).head(10)
print("\nüõë Venues with Most Batter Dismissals:\n", venue_dismissals)

# Top venues with high average SR (batting friendly)
plt.figure(figsize=(12, 5))
sns.barplot(x='venue', y='strike_rate', data=df_batter_venue, estimator='mean', ci=None, palette='viridis')
plt.xticks(rotation=45)
plt.title("üöÄ Average Batter Strike Rate per Venue")
plt.tight_layout()
plt.show()

# Best batters at top batting-friendly venues
top_sr_venues = df_batter_venue.groupby('venue')['strike_rate'].mean().sort_values(ascending=False).head(3).index.tolist()
for venue in top_sr_venues:
    print(f"\nüî• Top Batters at {venue}:\n")
    print(df_batter_venue[df_batter_venue['venue'] == venue].sort_values(by='strike_rate', ascending=False).head(5))

# Create venue-based batting score
df_batter_venue['venue_score'] = df_batter_venue['strike_rate'] / (df_batter_venue['dismissals'] + 1)
```

##  teams home and away

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df_team_venue = pd.read_csv("/content/teamwise_home_and_away.csv")

# Preview
print("üßæ Data Shape:", df_team_venue.shape)
print("\nüîç First 5 Rows:\n", df_team_venue.head())
print("\nüßº Missing Values:\n", df_team_venue.isnull().sum())
print("\nüìä Data Types:\n", df_team_venue.dtypes)

# Summary
print("\nüìà Descriptive Stats:\n", df_team_venue.describe())

# Sort and visualize
df_team_venue_sorted = df_team_venue.sort_values(by='home_win_percentage', ascending=False)

# Plot: Home vs Away win %
plt.figure(figsize=(12, 6))
bar_width = 0.4
teams = df_team_venue_sorted['team']
x = range(len(teams))

plt.bar(x, df_team_venue_sorted['home_win_percentage'], width=bar_width, label='Home Win %', color='green')
plt.bar([p + bar_width for p in x], df_team_venue_sorted['away_win_percentage'], width=bar_width, label='Away Win %', color='orange')

plt.xticks([p + bar_width/2 for p in x], teams, rotation=45)
plt.ylabel("Win Percentage")
plt.title("üìç Team-wise Home vs Away Win Percentage")
plt.legend()
plt.tight_layout()
plt.show()

# Create a team venue factor score
df_team_venue['home_advantage_score'] = df_team_venue['home_win_percentage'] - df_team_venue['away_win_percentage']
```
